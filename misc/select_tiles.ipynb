```python
def select_tiles_near_primaries(
    tiles_csv,
    primaries_csv,
    output_csv,
    max_sep_deg=1.0
):
    tiles = pd.read_csv(tiles_csv)
    primaries = pd.read_csv(primaries_csv)

    keep_mask = np.zeros(len(tiles), dtype=bool)

    # Loop over primaries (usually fewer than tiles)
    for p in primaries.itertuples(index=False):
        sep = angular_separation(
            tiles["ra"].values,
            tiles["dec"].values,
            p.ra,
            p.dec
        )
        keep_mask |= (sep <= max_sep_deg)

    filtered_tiles = tiles[keep_mask]

    # Save result
    filtered_tiles.to_csv(output_csv, index=False)

    print(f"Saved {len(filtered_tiles)} tiles to {output_csv}")

select_tiles_near_primaries(
    "cfis_r_tiles.csv",
    "kt17_primaries.csv",
    "kt17_containing_tiles2.csv",
    max_sep_deg=2.0
)
```
