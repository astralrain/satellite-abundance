```python
import pandas as pd
import numpy as np

def get_unique_tiles(files: list[str], output_file: str) -> pd.DataFrame:
    """
    Collect unique tile names from multiple input CSV files, files, and 
    write them to a new CSV filem output_file.

    Parameters:
    files (list[str]): List of paths to input CSV files. Each file must 
        contain a column 'tile_name'.
    output_file (str): Path to the output CSV file where the unique tile names 
        will be written.
    """
    all_tiles = set()

    for file in files:
        df = pd.read_csv(file)
        if "tile" not in df.columns:
            raise ValueError(f"{file} does not contain a 'tile' column.")
        all_tiles.update(df["tile"].unique())

    unique_tiles = pd.DataFrame({"tile": sorted(all_tiles)})
    unique_tiles.to_csv(output_file, index=False)

    return unique_tiles

csvs = ["kt17_matches_filtered.csv"]
get_unique_tiles(csvs, "unique_tiles_kt17_systems.csv")
```
